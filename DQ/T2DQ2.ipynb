One of the models I discussed in the first discussion question was the naive bayes classifier.

Bayes theorem is defined by the following equation:

\\[P(A|B)= \frac{P(B|A)P(A)}{P(B)}\\]


Where 

- \\(P(B|A)\\) is the likelihood
- \\(P(A)\\) is the prior probability
- \\(P(B)\\) is the predictor probability

It has an assumption of independence of the probabilities (what makes it naive) but in spite of this, it still works surprisingly well.

Below is an example using the iris dataset.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
# This is useful for normal classifications. There are others such as Bernoulli or multinomial.
from sklearn.naive_bayes import GaussianNB 
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
# load the data from the function
data = load_iris()
```

```python
# the iris function loads the data as a Bunch so we can place the data into dataframes to make
# handling the data easier.
X = data.data
y = data.target
```

```python
X.shape
```

```python
y.shape
```

```python
# Use the train_test_split function to make train and test sets.
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.2, random_state=69)
```

```python
# a crisp one-liner to do all the things
y_hat = GaussianNB().fit(X_train, y_train).predict(X_test)
```

```python
# How did we do?
print(accuracy_score(y_test, y_hat))
```

```python
# The orange line are the correct points
# the blue dots are the predicted points

plt.figure()
plt.plot(y_hat, '.')
plt.plot(y_test)
plt.xlabel('Index')
plt.ylabel('Predictions')
plt.title('Gaussian Naive Bayes on Iris')
plt.show()
```
