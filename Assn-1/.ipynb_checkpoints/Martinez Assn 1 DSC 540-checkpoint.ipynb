Because the variation is so different between home price, income, and the other
continuous variables, it's best to scale this data. The dependent variable doesn't need to be scaled because the model
will set the parameters to get the minimum cost function. We'd only need to scale y_test if y_train had been scaled.

The standard scaler function is:

\\[\frac{x_i - \mu}{\sigma}\\]

```python
X = data.drop('Price', axis=1)
y = data['Price']

# The standard scaler puts the data into a numpy array which is a bit less flexible than the dataframe.
scl = StandardScaler()
X = pd.DataFrame(scl.fit_transform(X))

# 20% test, 80% train
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size = 0.2,
                                                   random_state = 69)
```

## Model Building and Fitment

Now we will build a linear model. This will be the de facto ordinary least squares regression.

The function is as follows:

\\[ \hat{y} = \beta_0 + \beta_1 x_1 + \epsilon \\]

In order to ensure model accuracy, I'll use the coefficient of determination, or \\(R^2\\)

The function is as follows:

\\[SS_{tot} = \sum_{i}(y_i - \bar{y})^2\\]

\\[SS_{res} = \sum_{i} e^2_i\\]

\\[R^2 = 1 - \frac{SS_{res}}{SS_{tot}}\\]

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

```python
# These are the scaled coefs. While they are correct, they are a bit difficult to interpret.

model.coef_
```

```python
# We can unscale them for a more business-friendly explanation.

coefs2 = pd.DataFrame(model.coef_/scl.scale_)
coefs2
```

With our coefficients now computed, we can view the actual equation that we have built:

\\[\hat{y} = 1231277.0248333374 + 228853.64224356 x_1 + 164435.84671342 x_2 + 121054.20052859 x_3 + 1961.41011388 x _4 + 151825.60512927 x_5\\]

```python
coefficients = np.true_divide(model.coef_,  scl.scale_)
intercept = model.intercept_ - np.dot(coefficients, scl.mean_)
```

```python
# The intercept is what we expect mean price to be, or when X_i = 0. This is also scaled so we can also print the non-scaled

print(f'Scaled int: {model.intercept_}')
print(f'Raw int: {intercept}')
```

```python
y_hat = model.predict(X_test)
```

## Model Testing

First I'll do it programatically using the functions for MSE and R2.

```python
mse = mean_squared_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)

print(f'Mean Square error: {mse}')
print(f'R2 score: {r2}')
```

```python
print(f'RMSE Error: {sqrt(mse)}')
```

## Plot the predictions

```python
# plot the graph. This example refers to the income.

plt.scatter(X_test[0], y_test, color = "black")


plt.plot(X_test[0], y_hat, '.')
plt.xticks(())
plt.yticks(())
plt.show()
```

Here the \\(R^2\\) and mse can be calculated manually:

```python
SSR = ((y_test - y_hat) ** 2).sum()
SST = ((y_test - y_test.mean()) ** 2).sum()
```

```python
R_manual = 1-(SSR/SST)
```

```python
r2_adjusted = 1 - (1-R_manual)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```

```python
print(f'R-squared by hand: {R_manual}')
print(f'Adjusted R2: {r2_adjusted}')
```
